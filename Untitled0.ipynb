```python
%%writefile compiler.cpp
#include <iostream>
#include <string>
#include <vector>
#include <map>
#include <memory>
#include <regex>

using namespace std;

// ==========================================
// 1. LEXICAL ANALYSIS (Tokenizer)
// ==========================================
struct Token {
    string type;
    string value;
    int line;
};

vector<Token> tokenize(const string& code) {
    vector<Token> tokens;
    // Regex matches loops, IO, assignments, braces, and tracks newlines
    regex re(R"((for|while)|(if)|(cout|printf)|([a-zA-Z_]\w*\s*=)|(\{)|(\})|(\n))");
    sregex_iterator next(code.begin(), code.end(), re);
    sregex_iterator end;
    int current_line = 1;

    while (next != end) {
        smatch match = *next;
        if (match[1].matched) tokens.push_back({"LOOP", match.str(), current_line});
        else if (match[2].matched) tokens.push_back({"IF", match.str(), current_line});
        else if (match[3].matched) tokens.push_back({"IO", match.str(), current_line});
        else if (match[4].matched) tokens.push_back({"ASSIGN", match.str(), current_line});
        else if (match[5].matched) tokens.push_back({"LBRACE", "{", current_line});
        else if (match[6].matched) tokens.push_back({"RBRACE", "}", current_line});
        else if (match[7].matched) current_line++; // Increment line counter
        next++;
    }
    return tokens;
}

// ==========================================
// 2. SYNTAX ANALYSIS (AST Nodes)
// ==========================================
class AssignmentNode;
class IONode;
class LoopNode;
class BlockNode;

// The Visitor interface defines how we traverse our AST
class Visitor {
public:
    virtual void visit(AssignmentNode* node) = 0;
    virtual void visit(IONode* node) = 0;
    virtual void visit(LoopNode* node) = 0;
    virtual void visit(BlockNode* node) = 0;
};

// Base AST Node
class ASTNode {
public:
    int line;
    ASTNode(int l) : line(l) {}
    virtual ~ASTNode() = default;
    virtual void accept(Visitor& v) = 0;
};

// Specific Node Types
class AssignmentNode : public ASTNode {
public:
    AssignmentNode(int l) : ASTNode(l) {}
    void accept(Visitor& v) override { v.visit(this); }
};

class IONode : public ASTNode {
public:
    IONode(int l) : ASTNode(l) {}
    void accept(Visitor& v) override { v.visit(this); }
};

class BlockNode : public ASTNode {
public:
    vector<unique_ptr<ASTNode>> statements;
    BlockNode(int l) : ASTNode(l) {}
    void accept(Visitor& v) override { v.visit(this); }
};

class LoopNode : public ASTNode {
public:
    unique_ptr<BlockNode> body;
    LoopNode(int l) : ASTNode(l) {}
    void accept(Visitor& v) override { v.visit(this); }
};

// ==========================================
// 3. SYNTAX ANALYSIS (Parser)
// ==========================================
class Parser {
    vector<Token> tokens;
    size_t pos = 0;

public:
    Parser(vector<Token> t) : tokens(t) {}

    unique_ptr<BlockNode> parseBlock() {
        auto block = make_unique<BlockNode>(pos < tokens.size() ? tokens[pos].line : 0);
        while (pos < tokens.size()) {
            if (tokens[pos].type == "RBRACE") {
                break; // End of block
            }
            auto stmt = parseStatement();
            if (stmt) block->statements.push_back(move(stmt));
        }
        return block;
    }

    unique_ptr<ASTNode> parseStatement() {
        Token t = tokens[pos++];
        if (t.type == "LOOP") {
            auto loop = make_unique<LoopNode>(t.line);
            // Fast-forward to the block's opening brace to keep parsing simple
            while(pos < tokens.size() && tokens[pos].type != "LBRACE") pos++;
            if (pos < tokens.size() && tokens[pos].type == "LBRACE") {
                pos++; // Consume '{'
                loop->body = parseBlock();
                pos++; // Consume '}'
            }
            return loop;
        }
        else if (t.type == "IO") return make_unique<IONode>(t.line);
        else if (t.type == "ASSIGN") return make_unique<AssignmentNode>(t.line);

        return nullptr; // Ignore unknown/empty constructs for simplicity
    }
};

// ==========================================
// 4. SEMANTIC ANALYSIS (Energy Visitor)
// ==========================================
class EnergyVisitor : public Visitor {
public:
    map<int, int> line_costs;
    int current_multiplier = 1;

    void visit(AssignmentNode* node) override {
        // Base Assignment (1) + Memory Access (5) = 6
        int cost = (1 + 5) * current_multiplier;
        line_costs[node->line] += cost;
    }

    void visit(IONode* node) override {
        // I/O overhead (50)
        int cost = 50 * current_multiplier;
        line_costs[node->line] += cost;
    }

    void visit(BlockNode* node) override {
        for (auto& stmt : node->statements) {
            stmt->accept(*this);
        }
    }

    void visit(LoopNode* node) override {
        // Evaluate the inner body with a 10x multiplier
        int prev_multiplier = current_multiplier;
        current_multiplier *= 10;

        if (node->body) {
            node->body->accept(*this);
        }

        // Restore multiplier after loop finishes
        current_multiplier = prev_multiplier;
    }

    void printJSON() {
        cout << "{\n";
        int count = 0;
        for (auto const& [line, cost] : line_costs) {
            cout << "  \"line " << line << "\": " << cost;
            if (++count < line_costs.size()) cout << ",";
            cout << "\n";
        }
        cout << "}\n";
    }
};

// ==========================================
// MAIN EXECUTION
// ==========================================
int main() {
    // Input C++ code snippet (Notice the newlines for line-tracking)
    string code = R"(
for(int i=0; i<n; i++) {
    cout << i;
    a = 5;
}
    )";

    cout << "--- 1. Tokenizing Code ---\n";
    vector<Token> tokens = tokenize(code);

    cout << "--- 2. Building AST ---\n";
    Parser parser(tokens);
    auto ast = parser.parseBlock();

    cout << "--- 3. Semantic Analysis (Calculating Energy) ---\n";
    EnergyVisitor analyzer;
    ast->accept(analyzer);

    cout << "\n--- Result (JSON) ---\n";
    analyzer.printJSON();

    return 0;
}
```
